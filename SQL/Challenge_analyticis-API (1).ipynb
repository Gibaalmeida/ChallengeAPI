```python

import requests

# Defina os parâmetros e cabeçalhos
url = "https://api.mercadolibre.com/oauth/token"
headers = {
    "accept": "application/json",
    "content-type": "application/x-www-form-urlencoded"
}
data = {
    "grant_type": "authorization_code",
    "client_id": "4402280294489692",  # Seu Client_ID
    "client_secret": "x",  # Sua Secret_Key
    "code": "TG-67eb4297ef9da800016121e0-1826109381",  # Seu novo Authorization Code
    "redirect_uri": "https://www.google.com.br",  # Sua Redirect_URI registrada
    "code_verifier": "SEU_CODE_VERIFIER"  # Se aplicável
}

# Faça a requisição POST
response = requests.post(url, headers=headers, data=data)

# Verifique o resultado
if response.status_code == 200:
    print("Access Token:", response.json().get("access_token"))
else:
    print("Erro:", response.json())

```

```python
import requests
import csv

# O Access Token gerado com o TG
access_token = 'x'  

# Buscando tudo que aparecer de TV samsumg
url = 'https://api.mercadolibre.com/sites/MLB/search?q=TV+Samsung'


headers = {
    'Authorization': f'Bearer {access_token}'
}


response = requests.get(url, headers=headers)

# Verificar se a requisição foi bem-sucedida
if response.status_code == 200:
    # Resposta da API (dados dos produtos)
    data = response.json()

    # Caminho do arquivo CSV da minha maquina local
    csv_file = r'C:\Users\gilbe\OneDrive\Desktop\Mercado Livre\Case avanço internet\tvs_samsung.csv'

    # Abrindo o arquivo CSV em modo de escrita
    with open(csv_file, mode='w', newline='', encoding='utf-8') as file:
        writer = csv.writer(file)

        # Escrevendo o cabeçalho no CSV
        writer.writerow(['ID', 'Título', 'Preço', 'URL', 'Estado', 'Quantidade'])

        # Iterando sobre os itens retornados
        for item in data['results']:
            item_id = item.get('id')
            title = item.get('title')
            price = item.get('price')
            url = item.get('permalink')
            condition = item.get('condition')
            available_quantity = item.get('available_quantity')

            # Escrevendo as informações no CSV
            writer.writerow([item_id, title, price, url, condition, available_quantity])

    print(f"Dados salvos em: {csv_file}")
else:
    print(f"Erro ao consultar os produtos: {response.status_code} - {response.text}")

```
